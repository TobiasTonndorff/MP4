```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
from joblib import dump

# Assuming df contains your DataFrame with the dataset

# Splitting data into features (X) and target variable (y)
X = df.drop('Attrition', axis=1)
y = df['Attrition']

# Perform one-hot encoding for the 'EducationField' column
X_encoded = pd.get_dummies(X, columns=['EducationField'])

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Training Decision Tree classifier
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Training Naive Bayes classifier
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

# Predicting using Decision Tree classifier
dt_pred = dt_model.predict(X_test)

# Predicting using Naive Bayes classifier
nb_pred = nb_model.predict(X_test)

# Calculating accuracy of Decision Tree classifier
dt_accuracy = accuracy_score(y_test, dt_pred)
print('Decision Tree Accuracy:', dt_accuracy)

# Printing classification report for Decision Tree classifier
print("Decision Tree Classification Report:")
print(classification_report(y_test, dt_pred))

# Calculating accuracy of Naive Bayes classifier
nb_accuracy = accuracy_score(y_test, nb_pred)
print('Naive Bayes Accuracy:', nb_accuracy)

# Printing classification report for Naive Bayes classifier
print("Naive Bayes Classification Report:")
print(classification_report(y_test, nb_pred))


# Save Decision Tree model to a file
dump(dt_model, 'dt_model.joblib')

# Save Naive Bayes model to a file
dump(nb_model, 'nb_model.joblib')
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


# Assuming you have your employee data in a variable called 'employee_data'

# Step 2: Clustering with K-Means
def apply_kmeans(data, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(data)
    labels = kmeans.labels_
    return labels

# Step 3: Evaluate Clustering
def evaluate_clustering(data, max_clusters=10):
    silhouette_scores = []
    for n_clusters in range(2, max_clusters+1):
        labels = apply_kmeans(data, n_clusters)
        silhouette_score_value = silhouette_score(data, labels)
        silhouette_scores.append(silhouette_score_value)
    return silhouette_scores

# Step 4: Recommendation
def recommend_cluster_configuration(silhouette_scores):
    best_cluster_config = np.argmax(silhouette_scores) + 2  # Adding 2 to align with cluster count starting from 2
    best_score = np.max(silhouette_scores)
    return best_cluster_config, best_score

# Applying the functions
max_clusters_to_try = 10  # You can adjust this as per your needs
silhouette_scores = evaluate_clustering(df, max_clusters=max_clusters_to_try)
best_cluster_config, best_score = recommend_cluster_configuration(silhouette_scores)

print(f"The recommended cluster configuration is {best_cluster_config} clusters with a silhouette score of {best_score}.")
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
employee_data_pca = pca.fit_transform(df)

# Apply K-Means clustering with the recommended configuration (3 clusters)
labels = apply_kmeans(df, best_cluster_config)

# Visualize the clustering
plt.figure(figsize=(10, 6))
for cluster_label in range(best_cluster_config):
    cluster_data = employee_data_pca[labels == cluster_label]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], label=f'Cluster {cluster_label+1}')

plt.title('Clustering of Employees (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```
